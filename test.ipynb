```python
# Webスクレイピングに必要なライブラリインストール
import requests
from bs4 import BeautifulSoup
```

# HTTPリクエスト
## 基本のリクエスト

```python
#アクセスしたいサイトのurl
url = "https://www.musashino-u.ac.jp/"

#サーバーにリクエスト送信
#レスポンスを変数に格納
res = requests.get(url)
```

```python
res.encoding = res.apparent_encoding

print(f"レスポンス:{res}") #レスポンスオブジェクト
print(f"レスポンスの型:{type(res)}")#レスポンスオブジェクトの型
print(f"ステータスコード:{res.status_code}")#レスポンスオブジェクトの型
print(f"ステータスメッセージ:{res.reason}")#レスポンスオブジェクトの型

print(f"リクエスト:{res.request}")#リクエストオブジェクト
print(f"リクエスト:{type(res.request)}")#リクエストオブジェクトの型

print(f"リクエストヘッダー:{res.request.headers}")#リクエストヘッダー
print(f"リクエストメソッド:{res.request.method}")#リクエストヘッダー
print(f"リクエストURL:{res.request.url}")#リクエストヘッダー
```

```python

print(f"レスポンスヘッダー:{res.headers}")  # レスポンスヘッダー
print(f"レスポンスボディ:{res.text}")  # レスポンスボディ
```

```python
# 悪用できるやつ

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

```python
#
url = "https://spojoba.com/articles/354"
res = requests.get(url,headers=headers)
print(f"ステータスコード:{res.status_code}")
print(f"レスポンスボディ:{res.text}")
```

## パラメータ付きのリクエスト


```python
# url = "https://www.musashino-u.ac.jp/search.html?query=データサイエンス"

url = "https://www.musashino-u.ac.jp/search.html"
params = {
    "query" : "データサイエンス"
}
res = requests.get(url, headers=headers, params=params)

res.encoding = res.apparent_encoding

print(f"ステータスコード:{res.status_code}")
print(f"レスポンスボディ:{res.text}")
```

# BeautifulSoupでHTMLソースから任意の情報を取得する

```python
url = "https://www.musashino-u.ac.jp/"
res = requests.get(url,headers=headers)
res.encoding = res.apparent_encoding
print(f"ステータスコード：{res.status_code}")#
```

```python
# HTMLのソースコードを解析
soup = BeautifulSoup(res.text, "html.parser")
print(f"soupの型:{type(soup)}")
```

```python
soup_a = soup.a
print(f"<a>タグ:{soup_a}")
print(f"<a>タグの型:{type(soup_a)}")
```

```python
soup_div = soup.div
# print(f"<div>タグ:{soup_div}")
# print(f"<div>タグの型:{type(soup_div)}")

soup_div_div = soup_div.div
print(f"<div>タグの中の<div>タグ:{soup_div_div}")
print(f"<div>タグの中の<div>タグの型:{type(soup_div_div)}")
```

```python
# Tagオブジェクト.stringでタグ内のテキストを取得できる。
print(f"<a>タグのテキスト:{soup_a.string}")
print(f"<div>タグの中の<div>タグののテキスト:{soup_div_div.string}")
```
