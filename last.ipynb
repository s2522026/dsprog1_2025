```python
import requests
from bs4 import BeautifulSoup
import time

starturl = "https://www.musashino-u.ac.jp/"  # クロールを開始するURL（武蔵野大学のトップページ）
visited_links = set()  # 訪れたリンクを記録する
urls = {}  # URLとそのタイトルを格納する辞書

# リンクを再帰的にクロールする関数


def get_links(url):
    try:
        # 指定されたURLにリクエストを送信
        res = requests.get(url)
        res.raise_for_status()  # HTTPエラーが発生した場合は例外をスロー

        # レスポンスのHTMLコンテンツをBeautifulSoupでパース
        html_soup = BeautifulSoup(res.content, 'html.parser')

        # ページ内のすべての<a>タグを取得
        a_tags = html_soup.find_all('a')
        for a_tag in a_tags:
            href = a_tag.get('href')  # <a>タグのhref属性を取得
            if href:  # hrefが存在する場合
                # 内部リンクかどうかを判定
                if href.startswith('https://www.musashino-u.ac.jp/') or href.startswith('/'):
                    # 相対リンクを絶対URLに変換
                    if href.startswith('/'):
                        href = starturl + href  # 相対URLを基に完全なURLを作成

                    # 訪れたリンクでない場合
                    if href not in visited_links:
                        visited_links.add(href)  # 訪れたリンクとしてセットに追加

                        # リンクのタイトルを取得
                        try:
                            link_res = requests.get(href)  # リンク先にリクエストを送信
                            link_soup = BeautifulSoup(
                                link_res.content, 'html.parser')  # レスポンスをパース
                            title = link_soup.title.string.strip(
                            ) if link_soup.title else 'No title'  # ページのタイトルを取得

                            # 訪れたURLとタイトルを辞書に格納
                            if href not in urls:
                                urls[href] = title

                                # URLとタイトルを表示
                                print(f'{href}: {title}')
                        except Exception as e:
                            # エラーが発生した場合はエラーメッセージを表示
                            print(f'Error accessing {href}: {e}')

                        # サーバーに負担をかけないため、リクエスト間隔を設定
                        time.sleep(0.1)

                        # リンク先のページを再帰的にクロール
                        get_links(href)
    except Exception as e:
        print(f'Error accessing {url}: {e}')  # 最上位URLでエラーが発生した場合にエラーメッセージを表示


# クロールの開始
get_links(starturl)

# 収集したURLとタイトルを表示
print('Links and titles:')
for idx, (link, title) in enumerate(urls.items(), start=1):
    print(f'{idx}. {link}: {title}')

# URLとタイトルをファイルに保存
with open('links_and_titles.txt', 'w', encoding='utf-8') as f:
    for idx, (link, title) in enumerate(urls.items(), start=1):
        f.write(f'{link}\t{title}\n')
```
